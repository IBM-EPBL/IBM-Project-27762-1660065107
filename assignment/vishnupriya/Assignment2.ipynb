# ***6.Find the outliers and replace the outliers***

# *Finding outliers*

```python
import pandas as pd
import numpy as np
import plotly.express as px
df=pd.read_csv('Churn_Modelling.csv')
```

```python
df1= px.histogram(df, x='Age')

df1.show()
```

**finding using using box plot**

```python
df2 = px.box(df, y='Age')

df2.show()
```

**scatterplot for outliers**

```python
f1 = px.scatter(x=df['RowNumber'], y=df['Age'])

f1.show()
```

**statistical method**

```python
def find_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   outliers = df[((df<(q1-1.5*IQR)) | (df>(q3+1.5*IQR)))]

   return outliers
```

```python
outliers = find_outliers_IQR(df['Age'])

print('number of outliers: '+ str(len(outliers)))

print('max outlier value: '+ str(outliers.max()))

print('min outlier value: '+ str(outliers.min()))

outliers
```

```python
outliers = find_outliers_IQR(df[['RowNumber','Age']])
outliers
```

# **Replacing Outliers**

```python
def impute_outliers_IQR(df):

   q1=df.quantile(0.25)

   q3=df.quantile(0.75)

   IQR=q3-q1

   upper = df[~(df>(q3+1.5*IQR))].max()

   lower = df[~(df<(q1-1.5*IQR))].min()

   df = np.where(df > upper,

       df.mean(),

       np.where(

           df < lower,

           df.mean(),

           df

           )

       )

   return df
```

```python
df['Age'] = impute_outliers_IQR(df['Age'])

df.describe()['Age']
```
